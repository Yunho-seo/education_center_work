```python
# 이름, 국어, 영어, 수학 점수 3명분을 입력받아 
# 데이터 프레임으로 변환하고 합계와 평균 구하기 (과목별로, 학생별로 apply() 함수 적용)
# 데이터프레임을 합계 기준으로 정렬하고 등수를 입력
# 과목별 분산과 표준편차 구하기 (apply() 함수 적용)
# 과목별 상관계수와 공분산 구하기

# pymysql을 이용하여 DB에 있는 성적 데이터를 가지고 온 후 데이터 프레임을 만들고
# 과목 별 평균과 표준편차를 구하기
import numpy as np
import pandas as pd

names = []
kors = []
engs = []
mats = []

for i in range(3):
    name = input(f"{i+1}번째 학생의 이름: ")
    kor = int(input(f"{i+1}번째 학생의 국어점수: "))
    eng = int(input(f"{i+1}번째 학생의 영어점수: "))
    mat = int(input(f"{i+1}번째 학생의 수학점수: "))
    names.append(name)
    kors.append(kor)
    engs.append(eng)
    mats.append(mat)  # list


data = {'이름': names, '국어': kors, '영어': engs, '수학': mats}  # dict
df = pd.DataFrame(data)
```

```python
df
```

```python
# 파생변수 (합계와 평균)
# df['합계'] = df['국어'] + df['영어'] + df['수학']
# df['평균'] = round((df['국어'] + df['영어'] + df['수학']) / 3, 2)
# df

def total(row):
    return row['국어'] + row['영어'] + row['수학']

def avg(row):
    return round((row['국어'] + row['영어'] + row['수학']) / 3, 2)

df['총점'] = df.apply(total, axis=1)
df['평균'] = df.apply(avg, axis=1)
print(df)

```

```python
df
```

```python
# 합계 값으로 정렬하고 등수 매기기

df['등수'] = df['총점'].rank(method='min', ascending=False).astype(int)
print(df)
```

```python
df
```

```python
# 과목별로 apply() 함수 사용하여 분산, 표준편차 구하기
df['분산'] = df[['국어', '영어', '수학']].apply(np.var, axis=1)
df['표준편차'] = df[['국어', '영어', '수학']].apply(np.std, axis=1)

df
```

```python
# 과목별 상관계수(corr())와 공분산(cov()) 구하기
corr_matrix = df[['국어', '영어', '수학']].corr()
cov_matrix = df[['국어', '영어', '수학']].cov()

print("상관계수:", corr_matrix, '\n')
print("공분산:", cov_matrix, '\n')
```
