```python
# 문제
# sales 데이터를 로딩 
import pandas as pd

path = '/content/drive/MyDrive/sales.csv'
df = pd.read_csv(path)
df
```

```python
# 'Customer Number'를 int형으로 처리
df['Customer Number'] = df['Customer Number'].astype(int)
df
```

```python
# df['Jan Units']를 int형으로 처리 
df["Jan Units"] = pd.to_numeric(df["Jan Units"], errors='coerce').astype('Int64')
df
```

```python
# '2016' 데이터를 float형으로 처리하고 숫자 데이터에 들어있는 ,와 $ 제거
df['2016'] = df['2016'].replace('[\$,]', '', regex=True).astype(float)
df
```

```python
# 수치형 데이터 중 %를 수치로 바꾸기 (문자를 떼고, 100으로 나누기)
df['Percent Growth'] = df['Percent Growth'].str.rstrip('%').astype('float') / 100.0
df
```

```python
# Active를 True / False로 변경 (boolean 값)
df['Active'] = df['Active'].apply(lambda x: True if x == 'Y' else (False if x=='N' else None))
df
```
