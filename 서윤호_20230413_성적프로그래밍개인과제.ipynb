```python
def input_data():
    name = input("학생 이름을 입력하세요: ")
    kor = int(input("국어 점수를 입력하세요: "))
    mat = int(input("수학 점수를 입력하세요: "))
    eng = int(input("영어 점수를 입력하세요: "))
    return name, kor, mat, eng

def calc_avg(kor, mat, eng):
    avg = (kor + mat + eng) / 3
    return avg

def calc_grade(avg):
    if avg >= 90:
        grade = 'A'
    elif avg >= 80:
        grade = 'B'
    elif avg >= 70:
        grade = 'C'
    elif avg >= 60:
        grade = 'D'
    else:
        grade = 'F'
    return grade

def print_data(name, kor, mat, eng, avg, grade):
    print("학생 이름:", name)
    print("국어 점수:", kor)
    print("수학 점수:", mat)
    print("영어 점수:", eng)
    print("평균 점수:", avg)
    print("학점:", grade)

def calc_rank(avg, scores):
    rank = 1
    for score in scores:
        if avg < score:
            rank += 1
    return rank

def menu():
    scores = []
    while True:
        print("1. 입력")
        print("2. 계산")
        print("3. 학점계산")
        print("4. 출력")
        print("5. 등수")
        print("6. 종료")
        menu_num = int(input("메뉴를 선택하세요: "))
        
        if menu_num == 1:
            name, kor, mat, eng = input_data()
            scores.append(calc_avg(kor, mat, eng))
        elif menu_num == 2:
            if len(scores) == 0:
                print("데이터가 없다.")
            else:
                print("평균 점수:", sum(scores) / len(scores))
        elif menu_num == 3:
            if len(scores) == 0:
                print("데이터가 없다.")
            else:
                for i in range(len(scores)):
                    grade = calc_grade(scores[i])
                    print(f"{i+1}번째 학생의 학점은 {grade}입니다.")
        elif menu_num == 4:
            if len(scores) == 0:
                print("데이터가 없다.")
            else:
                for i in range(len(scores)):
                    name, kor, mat, eng = input_data()
                    avg = calc_avg(kor, mat, eng)
                    grade = calc_grade(avg)
                    print_data(name, kor, mat, eng, avg, grade)
        elif menu_num == 5:
            if len(scores) == 0:
                print("데이터가 없다.")
            else:
                avg = sum(scores) / len(scores)
                rank = calc_rank(avg, scores)
                print("등수는", rank, "등")
        elif menu_num == 6:
            print("종료")
            break
        else:
            print("잘못된 번호입니다.")

menu()
```
